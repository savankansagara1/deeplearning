```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
import pandas as pd

df = pd.DataFrame({
    "soil_moisture": [61, 24, 81, 70, 30, 45, 27, 66, 53, 29, 15, 78, 34, 22, 60, 48],
    "temperature_c": [35, 15, 26, 36, 26, 32, 28, 38, 31, 33, 20, 37, 25, 34, 29, 21],
    "sunlight_hours": [5, 7, 5, 9, 7, 8, 10, 6, 11, 4, 9, 3, 7, 8, 6, 10],
    # 1 = needs water, 0 = doesn't need water
    "needs_water": [0, 1, 0, 1, 0, 1, 1, 1, 1, 1, 1, 0, 0, 1, 0, 1]
})

print(df)
```

```python
 df.columns
```

```python
X = df[['soil_moisture', 'temperature_c', 'sunlight_hours']]
Y = df['needs_water']
```

```python
X_min = X.min()
X_max = X.max()
X_norm = (X - X_min) / (X_max - X_min + 1e-8)
```

```python
X_norm
```

```python
X_train, X_test, y_train, y_test = train_test_split( X_norm, Y, test_size=0.25, random_state=42)
```

```python
model = keras.Sequential([
    layers.Input(shape=(X_train.shape[1],)),
    layers.Dense(8,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer="sgd",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
history = model.fit(X_train,y_train,epochs=100,validation_data=(X_test,y_test),batch_size=4,verbose=1)
```
